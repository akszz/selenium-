```python
#libs
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time

#commands
driver = webdriver.Chrome(r"C:\Users\chromedriver.exe")
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver.get('http://www.flipkart.com/')
```

```python
search_sunglasses=driver.find_element_by_class_name("_3704LK")
search_sunglasses.send_keys("Sunglasses")
```

```python
search_button=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_button.click()
```

```python
Brands=[]
Descriptions=[]
prices=[]
```

```python
Brands_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
Brands_tags[0:100]
```

```python
for i in Brands_tags:
    Brand=i.text
    Brands.append(Brand)
Brands[0:100]
```

```python
Descriptions_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
Descriptions_tags[0:100]
```

```python
for i in Descriptions_tags:
    Description=i.text
    Descriptions.append(Description)
Descriptions[0:100]
```

```python
prices_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
prices_tags[0:100]
```

```python
for i in prices_tags:
    price=i.text
    prices.append(price)
prices[0:100]
```

```python
print(len(Brands),len(Descriptions),len(prices))
```

```python
sngl=pd.DataFrame({})
sngl['BRAND']=Brands[0:100]
sngl['DESCRIPTION']=Descriptions[0:100]
sngl['PRICE']=prices[0:100]
```
